```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path='Dataset .xlsx'
```

```python
data = pd.read_excel(path)
```

```python
prize_range_counts = data['Price range'].value_counts()
```

```python
plt.figure(figsize=(8,6))
prize_range_counts.plot(kind='bar',color='skyblue')
plt.title('distribution of prize range among resturants')
plt.xlabel('price range')
plt.ylabel('number of resturants')
plt.xticks(rotation=0)
plt.show()
```

```python
total_resturants = len(data)
```

```python
percentage_per_prize_range = (prize_range_counts / total_resturants) * 100
```
