```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path ='Dataset .xlsx'
```

```python
data = pd.read_excel(path)
```

```python
data
```

```python
rating_counts = data['Aggregate rating'].value_counts()
```

```python
most_common_rating = rating_counts.idxmax()
```

```python
print(f"the most common rating range is {most_common_rating}")
```

```python
plt.figure(figsize=(10,6))
plt.hist(data['Aggregate rating'],bins=10,color='skyblue',edgecolor='black')
plt.title('distribution of aggregate ratings')
plt.xlabel('aggregate rating')
plt.ylabel('number of resturants')
plt.show()
```

```python
average_votes = data['Votes'].mean()
```

```python
print(f"the average number of votes received by restauarnts is {average_votes }")
```
